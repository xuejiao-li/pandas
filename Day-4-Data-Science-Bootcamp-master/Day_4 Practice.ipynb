```python
# find the total departure delay for all the flight
departure_delay = flights['departure_delay']
departure_delay.sum()
```

```python
# the average of the delay for all the flight
departure_delay.mean()
```

```python
# compare the total departure delay time between each aire line, sort them from largest to lowest 
airline_delay = flights.groupby('airline').agg({'departure_delay':'sum'})
airline_delay.head()
```

```python
airline_delay
```

```python
# practice how to select with the index loc 
airline_delay.loc['AA', 'departure_delay']
```

```python
airline_delay.sort_values(by='departure_delay', ascending=False).head()
```

```python
# find the average of the delay for each airline
flight_delay_average = flights.groupby('airline').agg({'departure_delay':'mean'})
flight_delay_average.head()
```

```python
flight_delay_average.sort_values(by='departure_delay', ascending=False).head()
```

```python
# to check the type of the object returned by using groupby 
type(flight_delay_average)
```

```python
# reset the index
flight_delay_average.reset_index()
```
